```python
import CoolProp.CoolProp as cp

print("\nCOOLPROP CALCULATOR (SI UNITS)\n")

want = ["T", "P", "H", "S", "D", "Q", "V", "U", "W", "G", "A", "P_critical", "T_critical", "D_critical", "viscosity",
        "conductivity", "Prandtl", "Phase", "sigma"]

want_du = ["temperature[K]", "pressure[Pa]", "enthalpy[J/kg]", "entropy[J/kg.K]", "density[kg/m^3]", "quality[/]",
           "specific volume[m^3/kg]", "internal energy[J/kg]", "humidity ratio[water vapor/dry air]",
           "Gibbs free energy[J/kg]", "Helmholtz free energy[J/kg]", "critical pressure[Pa]", "critical temperature[K]",
           "critical density[kg/m^3]", "viscosity[Pa.s]",
           "thermal conductivity[W/m.K]", "Prandtl number[/]", "phase fraction[/]", "surface tension[N/m]"]

unit = ["[K]", "[Pa]", "[J/kg]", "[J/kg.K]", "[kg/m^3]", "[/]", "[m^3/kg]", "[J/kg]", "[water vapor/dry air]", "[J/kg]",
        "[J/kg]", "[Pa]", "[K]", "[kg/m^3]", "[Pa.s]", "[W/m.K]", "[/]", "[/]", "[N/m]"]

substance = ["1-Butene", "Acetone", "Air", "Ammonia", "Argon", "Benzene", "CarbonDioxide", "CarbonMonoxide",
             "CarbonylSulfide", "cis-2-Butene", "CycloHexane", "Cyclopentane", "CycloPropane", "D4", "D5", "D6",
             "Deuterium", "Dichloroethane", "DiethylEther", "DimethylCarbonate", "DimethylEther", "Ethane", "Ethanol",
             "EthylBenzene", "Ethylene", "EthyleneOxide", "Fluorine", "HeavyWater", "Helium", "HFE143m", "Hydrogen",
             "HydrogenChloride", "HydrogenSulfide", "IsoButane", "IsoButene", "Isohexane", "Isopentane", "Krypton",
             "m-Xylene", "MD2M", "MD3M", "MD4M", "MDM", "Methane", "Methanol", "MethylLinoleate", "MethylLinolenate",
             "MethylOleate", "MethylPalmitate", "MethylStearate", "MM", "n-Butane", "n-Decane", "n-Dodecane",
             "n-Heptane", "n-Hexane", "n-Nonane", "n-Octane", "n-Pentane", "n-Propane", "n-Undecane", "Neon",
             "Neopentane", "Nitrogen", "NitrousOxide", "Novec649", "o-Xylene", "OrthoDeuterium", "OrthoHydrogen",
             "Oxygen", "p-Xylene", "ParaDeuterium", "ParaHydrogen", "Propylene", "Propyne", "R11", "R113", "R114",
             "R115", "R116", "R12", "R123", "R1233zd(E)", "R1234yf", "R1234ze(E)", "R1234ze(Z)", "R124", "R1243zf",
             "R125", "R13", "R134a", "R13I1", "R14", "R141b", "R142b", "R143a", "R152A", "R161", "R21", "R218", "R22",
             "R227EA", "R23", "R236EA", "R236FA", "R245ca", "R245fa", "R32", "R365MFC", "R40", "R404A", "R407C", "R41",
             "R410A", "R507A", "RC318", "SES36", "SulfurDioxide", "SulfurHexafluoride", "Toluene", "trans-2-Butene",
             "Water", "Xenon"]

inp1 = input("I want to know: ")
if inp1 in want:
    if inp1 == "T_critical":
        a = want.index(inp1)
        b = want_du[a]
        u = unit[a]
        inp2 = input("for the substance: ")
        if inp2 in substance:
            answer = cp.PropsSI(inp1, inp2)
            print("\nthe " + b + " of " + str(inp2) + " = " + str(round(answer, 3)) + u)
        else:
            print("SUBSTANCE IS NOT APPLICABLE")
    elif inp1 == "P_critical":
        a = want.index(inp1)
        b = want_du[a]
        u = unit[a]
        inp2 = input("for the substance: ")
        if inp2 in substance:
            tc = cp.PropsSI("T_critical", inp2)
            inp3 = input("for a given: ")
            if inp3 in want:
                c = want.index(inp3)
                d = want_du[c]
                inp33 = input("at " + d + " = ")
                answer = cp.PropsSI("P", "T", tc, inp3, float(inp33), inp2)
                print("\nthe " + b + " of " + str(inp2) + " at the given (" + inp3 + ") = " + str(round(answer, 3)) + u)
            else:
                print("INPUT DOES NOT EXIST")
        else:
            print("SUBSTANCE IS NOT APPLICABLE")
    elif inp1 == "D_critical":
        a = want.index(inp1)
        b = want_du[a]
        u = unit[a]
        inp2 = input("for the substance: ")
        if inp2 in substance:
            tc = cp.PropsSI("T_critical", inp2)
            inp3 = input("for a given: ")
            if inp3 in want:
                c = want.index(inp3)
                d = want_du[c]
                inp33 = input("at " + d + " = ")
                answer = cp.PropsSI("D", "T", tc, inp3, float(inp33), inp2)
                print("\nthe " + b + " of " + str(inp2) + " at the given (" + inp3 + ") = " + str(round(answer, 3)) + u)
            else:
                print("INPUT DOES NOT EXIST")
        else:
            print("SUBSTANCE IS NOT APPLICABLE")
    elif inp1 != ("P_critical" or "T_critical" or "D_critical"):
        a = want.index(inp1)
        b = want_du[a]
        u = unit[a]
        inp2 = input("for a given: ")
        if inp2 in want:
            c = want.index(inp2)
            d = want_du[c]
            inp22 = input("at " + d + " = ")
            inp3 = input("for a given: ")
            if inp3 in want:
                e = want.index(inp3)
                f = want_du[e]
                inp33 = input("at " + f + " = ")
                inp4 = input("for the substance: ")
                if inp4 in substance:
                    answer = cp.PropsSI(inp1, inp2, float(inp22), inp3, float(inp33), inp4)
                    print("\nthe " + b + " of " + str(inp4) + " at the given (" + inp2 + ") and (" + inp3 + ") = " + str(round(answer, 3)) + u)
                else:
                    print("SUBSTANCE IS NOT APPLICABLE")
            else:
                print("INPUT DOES NOT EXIST")
        else:
            print("INPUT DOES NOT EXIST")
    else:
        print("INPUT DOES NOT EXIST")
else:
    print("INPUT DOES NOT EXIST")
```
